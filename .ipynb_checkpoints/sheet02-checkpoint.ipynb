```python
from scipy.cluster.hierarchy import dendrogram

def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack(
        [model.children_, model.distances_, counts]
    ).astype(float)
    
    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)
```

```python
plot_dendrogram(model, truncate_mode="level", p=86)
plt.xlabel("Number of points in node (or index of point if no parenthesis).")
plt.show()
```

#### scipy

```python
import scipy.cluster.hierarchy as sch
```

```python
linkage_matrix=sch.linkage(X,method="ward") 
```

```python
def get_labels(lab_type="std",stbl=stbl):
    lab_types={"std":("sample_type","sample_subtype"),"type":("sample_type",),"sub_type":("sample_subtype",)}
    labels=[]
    for sname in list(X.index):
        lab=stbl[stbl.name==sname].loc[:,lab_types[lab_type]]
        if len(lab_types[lab_type])>1:
            lab="_".join(list(lab.iloc[0]))
            labels.append(lab)
        else:
            lab=list(lab.iloc[0])
            labels.extend(lab)
    return labels 
```

```python
labels=get_labels(lab_type="std")
```

```python
labels[:10]
```

```python
fig,ax=plt.subplots(figsize=(10,5))

sch.dendrogram(linkage_matrix,ax=ax,labels=labels)
ax.set_title("linkage: ward")

fig.tight_layout()
plt.savefig(figures+"dendrogram_ward.png")
```

Let's wrap everything inside a function

```python
def cluster_plot(X,method="ward",lab_type="std",stbl=stbl,ax=None):
    
    if ax is None:
        fig,ax=plt.subplots(figsize=(10,5))

    linkage_matrix=sch.linkage(X,method=method) 
    
    labels=get_labels(stbl=stbl,lab_type=lab_type)
    
    sch.dendrogram(linkage_matrix,ax=ax,labels=labels)
    ax.set_title(f"linkage: {method}")
    
```

```python
fig,axs=plt.subplots(figsize=(10,15),nrows=3)

cluster_plot(X,method="single",ax=axs[0])
cluster_plot(X,method="complete",ax=axs[1])
cluster_plot(X,method="average",ax=axs[2])


fig.tight_layout()
plt.savefig(figures+"dendrograms_stdlabs.png")
```

```python
fig,axs=plt.subplots(figsize=(10,15),nrows=3)

cluster_plot(X,method="single",lab_type="type",ax=axs[0])
cluster_plot(X,method="complete",lab_type="type",ax=axs[1])
cluster_plot(X,method="average",lab_type="type",ax=axs[2])


fig.tight_layout()
plt.savefig(figures+"dendrograms_typelabs.png")
```

```python
fig,axs=plt.subplots(figsize=(10,15),nrows=3)

cluster_plot(X,method="single",lab_type="sub_type",ax=axs[0])
cluster_plot(X,method="complete",lab_type="sub_type",ax=axs[1])
cluster_plot(X,method="average",lab_type="sub_type",ax=axs[2])


fig.tight_layout()
plt.savefig(figures+"dendrograms_subtypelabs.png")
```
